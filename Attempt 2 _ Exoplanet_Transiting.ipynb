## Equations:
$\mathbf{\text{Apparent Magnitude:}}$
$$m = \frac{L}{4\pi d^{2}}$$
$$\text{m: apparent brightness :: }\frac{W}{m^{2}}$$
$$\text{L : Luminosity :: }W$$
$$\text{d : Distance :: }m$$
&emsp; <br>

$\mathbf{\text{Luminosity:}}$
$$L = 4\pi d^{2} m$$
$$\text{L : Luminosity :: }W$$
$$\text{d : Distance :: }m$$
$$\text{m: apparent brightness :: }\frac{W}{m^{2}}$$
&emsp; <br>

$$\text{-------------------------------------------------------------}$$
$\mathbf{\text{Bolometric Magnitude}}$
$$ \frac{L_{\odot}}{L_{\star}}=10^{0.4\left(M_{bol,\odot}-M_{bol,\star}\right)}$$

$$\text{-------------------------------------------------------------}$$
$\mathbf{\text{Apparent Magnitude}}$
$$100^{\frac{m-M}{5}}=\frac{F_{10}}{F} = \left(\frac{d}{10pc}\right)^{2}$$
$$M = m-5\log_{10}\left(d_{pc}\right) + 5 = m - 5\left(\log_{10}d_{pc}-1\right)$$
$$m = -2.5\log_{10}\left(\frac{L}{L_0}\right)$$

$$\text{-------------------------------------------------------------}$$
$\mathbf{\text{Stefan's Law, Luminosity:}}$
$$L = \sigma A T^{4}$$
$$L = 4\pi R^{2}\sigma T^{4}$$
$$\text{L : Luminosity :: }W$$
$$\sigma \text{ : Stefan-Boltzmann Constant :: }\frac{W}{m^{2}K^{4}}$$
$$\text{A : Surface Area of Star :: }m^{2}$$ 
$$\text{T : Temperature :: }K$$
$$\text{R : Radius of Star :: }m$$

$$\text{-------------------------------------------------------------}$$
$\mathbf{\text{Light Gathering Power:}}$
$$A = \pi r^{2}$$
$$\text{A : Area :: } m^{2}$$
$$\text{r : Radius of Telescope :: }m$$

$$\text{-------------------------------------------------------------}$$
$\mathbf{\text{Combination:}}$
$$m = \frac{R^{2}\sigma T^{4}}{d^{2}}$$
$$\text{m: apparent brightness :: }\frac{W}{m^{2}}$$
$$\text{R : Radius of Star :: }m$$
$$\text{Stefan-Boltzmann Constant :: 5.670374419 x }10^{-8}\frac{W}{m^{2}K^{4}}$$
$$\text{T :Temperature :: }K$$
$$\text{d : Distance :: }m$$

$$\text{-------------------------------------------------------------}$$
$\mathbf{\text{Calculating Chords of Circles as a Function of Angle, Radius, and Distance:}}$
$$\left(x-d\right)^{2} + y^{2} = R^{2}$$
$$y = x tan\theta$$
$$x = \frac{-b\pm \sqrt{b^{2}-4ac}}{2a}$$
$$0 = x^{2}\left(1+tan^{2}\theta\right)-2xd+d^{2}-R^{2}$$
$$x_{\begin{matrix}x_1,x_2\\+,-\end{matrix}}=\frac{d\pm\sqrt{d^2-\left(1+tan^{2}\theta\right)\left(d^{2}-R^{2}\right)}}{1+tan^{2}\theta}$$
$$y_{\begin{matrix}y_1,y_2\end{matrix}}=\sqrt{R^{2}-\left(x_{\begin{matrix}x_1,x_2\end{matrix}}-d\right)^{2}}$$
$$\text{Chord Vector: }\left<x_{2}-x_{1},y_{2}-y_{1}\right>$$
$$\text{Chord Length: }\sqrt{\left(x_{2}-x_{1}\right)^{2}+\left(y_{2}-y_{1}\right)^{2}}$$

$$\text{-------------------------------------------------------------}$$
$\mathbf{\text{Calculating Chords of Hollow Circles as a Function of Angle, Inner Radius, Outer Radius, and Distance:}}$
$$\left(x-d\right)^{2} + y^{2} = R^{2}$$
$$\left(x-d\right)^{2} + y^{2} = r^{2}$$
$$y = x tan\theta$$
$$x = \frac{-b\pm \sqrt{b^{2}-4ac}}{2a}$$
$$0 = x^{2}\left(1+tan^{2}\theta\right)-2xd+d^{2}-R^{2}$$
$$0 = x^{2}\left(1+tan^{2}\theta\right)-2xd+d^{2}-r^{2}$$
$$x_{\begin{matrix}x_1,x_4\\+,-\end{matrix}}=\frac{d\pm\sqrt{d^2-\left(1+tan^{2}\theta\right)\left(d^{2}-R^{2}\right)}}{1+tan^{2}\theta}$$
$$x_{\begin{matrix}x_2,x_3\\+,-\end{matrix}}=\frac{d\pm\sqrt{d^2-\left(1+tan^{2}\theta\right)\left(d^{2}-r^{2}\right)}}{1+tan^{2}\theta}$$
$$y_{\begin{matrix}y_1,y_4\end{matrix}}=\sqrt{R^{2}-\left(x_{\begin{matrix}x_1,x_4\end{matrix}}-d\right)^{2}}$$
$$y_{\begin{matrix}y_2,y_3\end{matrix}}=\sqrt{r^{2}-\left(x_{\begin{matrix}x_2,x_3\end{matrix}}-d\right)^{2}}$$
$$\text{Chord Vector: }\left<x_{4}-x_{3}+x_{2}-x_{1},y_{4}-y_{3}+y_{2}-y_{1}\right>$$
$$\text{Chord Length: }\sqrt{\left(x_{4}-x_{3}\right)^{2}+\left(y_{4}-y_{3}\right)^{2}} - \sqrt{\left(x_{2}-x_{1}\right)^{2}+\left(y_{2}-y_{1}\right)^{2}}$$

$$\text{-------------------------------------------------------------}$$
$\mathbf{\text{Calculating Intensity Based On Chord Length:}}$
$$ Intensity Chord = \left(1-a\right)^{Chord Length/\Delta Height}\times Intensity Array$$
